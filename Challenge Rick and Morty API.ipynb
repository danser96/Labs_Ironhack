```python
df=pd.DataFrame()
for i in range (2,126):
    url= f'https://rickandmortyapi.com/api/location/{i}'
    response = requests.get(url)
    results = response.json() 
    
    flattened_data = json_normalize(results)
    
    df = df.append(flattened_data)
    i += 1

df
```
