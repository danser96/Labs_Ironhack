20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

```python
url2 = 'https://www.emsc-csem.org/Earthquake/'
```

```python
page = requests.get(url2).content
```

```python
soup2 = BeautifulSoup(page, 'html.parser')
soup2
```

```python
earth_table = soup2.find_all("table")[3]#put the index to get it in tab form
earth_table
```

```python
table_heads = earth_table.find_all('th')
table_heads
```

```python
table_h = [i.text for i in table_heads]
table_h[1:]
```

```python
rows = earth_table.find("tbody").find_all("tr")
rows
```

```python
import pandas as pd
```

```python
l = []
for tr in rows:
    td = tr.find_all('td')
    row = [str(tr.get_text()).strip() for tr in td]
    l.append(row[3:])
print(l)
```

```python

df = pd.DataFrame(l)
df.head(10)
```

```python
df[1] = df[1]+df[2]
df.drop(2, inplace=True, axis=1)
```

```python
df[3] = df[3]+df[4]
df.drop(4, inplace=True, axis=1)
```

```python
df
```

```python
df.head(21)
```

```python
df.columns = table_h
df
```

```python
table_heads = lang_table.find_all('th')
table_heads
```

```python
rows = lang_table.find("tbody").find_all("tr")
rows
```

```python
row_lang = [i.text.strip().split('\n\n') for i in rows][0]
row_lang
```
