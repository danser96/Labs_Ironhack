```python
url2= 'https://stephaniestorey.com/blog/top-10-must-see-art-in-the-united-states'
```

```python
page2 = requests.get(url2).content
```

```python
paint_soup = BeautifulSoup(page2, 'html.parser')
paint_soup
```

```python
paragraph2 = paint_soup.find_all("p")
paragraph2
```

```python
for paragraph in paragraph2:
    print(paragraph.prettify())
```

```python
titles = paint_soup.find_all('p',{'class':''})
titles
```

```python
facts = [i.text for i in titles]
facts
```

# 10 must see Art pieces in USA

```python
df= pd.DataFrame(facts)
df[2:22]
```

```python
result = df.groupby(np.arange(len(df))//2).sum()
result[1:11]

with option_context('display.max_colwidth', 400):
    display(result[1:11])
```
