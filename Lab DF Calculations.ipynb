If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names.
```python
pokemon['Name'] = pokemon['Name'].str.replace('\w+Mega','Mega')
pokemon.head()
```

Create a new column called Combo Type whose value combines Type 1 and Type 2.
Rules:

If both Type 1 and Type 2 have valid values, the Combo Type value should contain both values in the form of <Type 1> <Type 2>. For example, if Type 1 value is Grass and Type 2 value is Poison, Combo Type will be Grass-Poison.

If Type 1 has valid value but Type 2 is not, Combo Type will be the same as Type 1. For example, if Type 1 is Fire whereas Type 2 is NaN, Combo Type will be Fire.

```python
pokemon['Combo Type'] = np.where(pokemon['Type 2'].isnull(), pokemon['Type 1'], 
                                 pokemon ['Type 1'] + '-' + pokemon ['Type 2'])
pokemon.head()
```

The second group is Preferred Customers whose aggregated expenses are between the 75th and 95th percentile.

Sub Problem 1: How to aggregate the amount_spent for unique customers?

Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?

Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

```python
sum_amount_spent = orders.groupby('CustomerID').sum()['amount_spent']
orders_df = pd.DataFrame(sum_amount_spent)
orders_df
```

```python
sum_amount_spent.quantile(0.95)
```

```python
sum_amount_spent.quantile(0.75)
```

```python
orders_df['Type of Customer'] = np.where((sum_amount_spent > sum_amount_spent.quantile(0.75)) & 
                                  (sum_amount_spent <= sum_amount_spent.quantile(0.95)), 
                                                'Preferred', 
                                  (np.where(sum_amount_spent > sum_amount_spent.quantile(0.95), 
                                                'VIP', 'Normal')))
```
